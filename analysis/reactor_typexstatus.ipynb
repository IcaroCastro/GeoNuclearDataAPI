```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('./data/data.csv')

agrupado = df.groupby(['reactor_type', 'status'])
display(agrupado.size())

reactor_types = df['reactor_type'].unique()
status = df['status'].unique()

# Compare the number of reactors by type and status
for reactor_type in reactor_types:
    df_reactor_type = df[df['reactor_type'] == reactor_type]
    x = df_reactor_type['status'].unique()
    y = [df_reactor_type[df_reactor_type['status'] == status]['status'].count() for status in x]
    
    plt.bar(x, y)
    plt.title(reactor_type)
    plt.xticks(rotation=-90)
    plt.show()
```
